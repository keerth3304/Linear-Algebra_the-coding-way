```python
importimport numpy as np # array operations
```

```python
p = [
    [0, 1/2, 0, 0, 0, 0, 0],
    [0, 0, 1/3, 0, 1/2, 0, 0],
    [1, 0, 0, 0, 0, 1/3, 0],
    [0, 0, 1/3, 1, 0, 0, 0],
    [0, 1/2, 0, 0, 0, 1/3, 0],
    [0, 0, 1/3, 0, 1/2, 0, 0],
    [0, 0, 0, 0, 0, 1/3, 1]
    ]
```

```python
n=len(p) # number of pages
c=[] # list to store the pages with no outgoing links
for i in range(len(p)):
    if(p[i][i]==1):
        c.append(i) # storing the index of the pages with no outgoing links

for i in range(n):
    for j in range(n):
        if(j in c): # checking if the page has no outgoing links
            p[i][j]=1/n # replacing the pages with no outgoing links with the pages with equal probability
p=np.array(p) # converting the list to array
```

```python
def google_matrix(p,alpha): # function to calculate the google matrix
    """This function calculates the google matrix.

    Parameters:
        p(matrix): The matrix with the probability of the pages.
        alpha(float): The probability.
    """
    n=len(p) # number of pages
    k= np.full((n, n), 1/n) # matrix with equal probability
    return alpha*p + (1-alpha)*k # returning the google matrix

probability=0.8
matrix=google_matrix(p,probability) # calling the function with the matrices and probability
```

```python
def page_rank(A):
    """This function calculates the page rank of the pages using the adjacency matrix.

    Parameters:
        A(matrix): The adjacency matrix of the pages.
    """
    eigenvalues, eigenvectors = np.linalg.eig(A) # eigenvalues and eigenvectors
    Index=0 # index of the eigenvector corresponding to eigenvalue
    for i in eigenvalues:
        if(round(i.real,2)==1):# checking if the eigenvalue is 1
            print("Eigenvector corresponding to eigenvalue 1 is: ", eigenvectors[:,Index])
            significant = eigenvectors[:,Index] # storing the eigenvector
            break
        Index+=1 # incrementing the index
    significant=significant.real/sum(significant.real) # scaling the significant vector
    significant=significant.round(3) # rounding off to 3 decimal places
    print("The page rank of the pages are: ")   # printing the page rank of the pages
    for i in range(len(significant)):
        print("Page{} : {}".format(i+1,significant[i]))
page_rank(matrix) # calling the function
```
