If a vector v and it's transformation A are given ,

Then v is considered to be the eigen vector of A if there exits some lambda.

A X v=lambda X v

```python
#Import the necessay libraries
import numpy as np #for defining array
import matplotlib.pyplot as plt #For plotting vectors
```

```python
A=np.array([[1,6],[5,2]]) #Define a two dimentional vector for A

#Define the vectors u and v
u=np.array([6,-5])
v=np.array([3,-2])
```

```python
# Compute the transformations Au and Av
Au=A@u #Multiply A and u vector for transformation
Av=A@v #Multiply A and v vector for transformation

print("Initial u Vector : ",u)  #Printing u vector
print("Transformed u vector :",Au)  #Printing Au vector
print("Initial v Vector : ",v)  #Printing v vector
print("Transformed v vector :",Av)  #Printing Av vector
```

```python
# Setting up the figure and specifying the figure size
plt.figure(figsize=(8, 6))

# Plotting vector u from origin to its coordinates, labeled 'u' in red
plt.plot([0, u[0]], [0, u[1]], 'red', label='u')

# Plotting vector v from origin to its coordinates, labeled 'v' in green
plt.plot([0, v[0]], [0, v[1]], 'Green', label='v')

# Plotting the transformed vector Au from origin to its coordinates, labeled 'Au' in yellow
plt.plot([0, Au[0]], [0, Au[1]], 'yellow', label='Au')

# Plotting the transformed vector Av from origin to its coordinates, labeled 'Av' in blue
plt.plot([0, Av[0]], [0, Av[1]], 'Blue', label='Av')

# Adding a horizontal line at y=0 and a vertical line at x=0 to indicate the axes, both in black
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Adding a grid to the plot for better visibility of vector positions
plt.grid(True)

# Setting labels for the x and y axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Setting a title for the plot
plt.title('Vector Transformations')

# Adding a legend to the plot to identify each vector
plt.legend()

# Displaying the plot
plt.show()
```
