To find the independence of eigenvectors :


*   All the eigenvectors are converted to matrix
*   Reduced Row echelon form is computed for the matrix
*   No.of Non zero rows are found . If it is found to be equal to no.of Non zero rows in initial matrix , Eigen vectors are said to be independent  .

Sympy module can be used to find the no.of non zero rows in a matrix and it's respective reduced row echelon form .




```python
#import necessary modules
import numpy as np
import sympy as sp
```

```python
A = np.array([[2,1,0],[1,3,-1],[0,-1,4]])  #Matrix is defined
```

```python
eigenvalues, eigenvectors = np.linalg.eig(A)

# Display the results
print("Eigenvalues of A:", eigenvalues)
print("Eigenvectors of A:\n", eigenvectors)
```

```python
ev=sp.Matrix(eigenvectors)
reduced_martix=ev.rref()   #reduced row echelon form is computed
print("Reduced Row echlon form of eigen vectors : \n",reduced_martix[0])
```

```python
def nonzero_rows(A):
  A=sp.Matrix(A)
  nonzero_indices = A.rowspace()
  #Number of non-zero rows are returned
  return len(nonzero_indices)

print("No.of NonZero rows in rref of eigen vector : ",nonzero_rows(reduced_martix[0]))
print("No.of NonZero rows in initial matrix A : ",nonzero_rows(A))
```
